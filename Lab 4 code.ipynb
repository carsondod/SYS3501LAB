```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
#20
min_sepal_width_virginica = iris[iris['species'] == 'virginica']['sepal_width'].min()
min_sepal_width_virginica
```

```python
#21
mean_petal_length = iris['petal_length'].mean()

filtered_df = iris[iris['petal_length'] > mean_petal_length]

num_observations = filtered_df.shape[0]
num_observations
```

```python
#22
versicolor_num = (iris['species'] == 'versicolor').sum()
print(versicolor_num)

sepal_length_mean = iris['sepal_length'].mean()
print(sepal_length_mean)

unique_species = iris['species'].nunique()
print(unique_species)

virginica_count = (iris['species'] == 'virginica').sum()
print(virginica_count)
```

```python
#23
correlation_matrix = iris.drop(columns='species').corr()
highest_corr_pair = correlation_matrix.unstack().sort_values(ascending=False)
highest_corr_pair = highest_corr_pair[highest_corr_pair < 1].idxmax()
highest_corr_pair
```

```python
#24
setosa_petal_length_mean = iris[iris['species'] == 'setosa']['petal_length'].mean()
print(setosa_petal_length_mean)

largest_sepal_width_species = iris.groupby('species')['sepal_width'].mean().idxmax()
print(largest_sepal_width_species)

setosa_sepal_length_std = iris[iris['species'] == 'setosa']['sepal_length'].std()
print(setosa_sepal_length_std)

largest_sepal_length_species = iris.groupby('species')['sepal_length'].mean().idxmax()
print(largest_sepal_length_species)
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
#25
second_class_survived = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)]
count = len(second_class_survived)

print(count)
```

```python
#26
first_class_not_survived = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]
min_age = first_class_not_survived['age'].min()
max_age = first_class_not_survived['age'].max()

print(min_age)
print(max_age)
```

```python
#27
total_females = len(titanic[titanic['sex'] == 'female'])
female_survived = len(titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)])

percentage_female_survived = (female_survived / total_females) *100

print(percentage_female_survived)
```

```python
#28
missing_values = titanic[['sex', 'embarked', 'deck', 'age']].isnull().sum()

print(missing_values)
```

```python
#29
survival_by_class = titanic.groupby('pclass')['survived'].mean()
print(survival_by_class)

survival_southampton = titanic[titanic['embarked'] == 'S']['survived'].mean()
print("Southampton survival rate: " + str(survival_southampton))

survival_cherbourg = titanic[titanic['embarked'] == 'C']['survived'].mean()
print("Cherbourg survival rate: " + str(survival_cherbourg))

corr_class = titanic['survived'].corr(titanic['pclass'])
print("Survival passenger class correlation: " + str(corr_class))

corr_fare = titanic['survived'].corr(titanic['fare'])
print("Cherbourg survival rate: " + str(corr_fare))
```
